# The Central Limit Theorum
In essence, the CLT states that the normal sampling distribution of XÂ¯ which holds exactly for
random samples from a normal distribution, also holds approximately for random samples from
nearly any distribution. 


```python
import matplotlib.pyplot as plt
import random as rn
from collections import Counter
```

```python
ls=[1,2,3,4,5,6]  #Outcomes of a dice
y=[]
```

## Probability distribution of the experiment : Outcomes of a roll of dice
The probability of any of the outcomes is equally likely hence every outcome has a probability attached to it as 1/6

```python
Outcomes = [i for i in range(1,7)]
Probability = [1/6]*6
plt.plot(Outcomes,Probability)
```

```python
for i in range(100000):
    ch=rn.choices(ls,k=4) # returns a random sample of size 4
    y.append(sum(ch)/4)   # 1000 such samples are taken
    # the avg value of the samples is stored (Xbar).
    # The sample distribution (distribution of X bar) follows a normal distribution
```

```python
    rc=Counter(y)  #counting the freqyency of each element
    print(rc)
```

```python
xc=[]
yc=[]

for i in rc:
    xc.append(i)
xc=sorted(xc)

for i in xc:
    yc.append(rc[i])
```

```python
plt.plot(xc,yc)
```
